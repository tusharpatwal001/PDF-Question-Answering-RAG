```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


def summarize_pdf_2(file_path):
    loader = PyPDFLoader(file_path)
    docs = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=100, chunk_overlap=20)
    splitted_text = text_splitter.split_documents(docs)
    llm = ChatGroq(model='gemma2-9b-it', temperature=0, api_key=GROQ_API_KEY)
    chain = load_summarize_chain(llm, chain_type="map_reduce")
    summary = chain.invoke(splitted_text)

    return summary
```

```python
summary = summarize_pdf_2("Illustrated_Transformer.pdf")

print("Summary: \n")
print(summary)
```

```python
from langchain_ollama import ChatOllama


def summarize_pdf_3(file_path):
    loader = PyPDFLoader(file_path)
    docs = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=100, chunk_overlap=20)
    splitted_text = text_splitter.split_documents(docs)
    llm = ChatOllama(model="gemma3:1b")
    chain = load_summarize_chain(llm, chain_type="map_reduce")
    summary = chain.invoke(splitted_text)

    return summary

summary = summarize_pdf_3("Illustrated_Transformer.pdf")

print("Summary: \n")
print(summary)
```
